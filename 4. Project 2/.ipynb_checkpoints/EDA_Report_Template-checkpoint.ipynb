## Selected Problem Statement: 
A venture capital firm struggles to reliably select seed-stage startups that will eventually secure follow-on funding, resulting in a portfolio of stalled companies. The Investment Analytics team needs to analyze historical funding data across industry categories, funding round counts, time between rounds, and geographic regions to identify the patterns most strongly associated with startups that successfully raise additional capital. Uncovering these historical trajectories will allow the firm to accurately refine its $10M seed-stage investment criteria and target startups with the highest odds of advancing.

## 1. Introduction & Objectives
Seed investing is a lot like placing early bets on raw ideas. For a venture capital firm, the goal isn't just to hand out cash; it's to find startups that can take that initial spark, build real momentum, and successfully unlock their next major round of funding.

The reality, though, is tough. A huge number of startups raise a seed check, burn through the money, and completely stall out. When you're managing a $10M fund, you can't afford to just guess or rely on a "gut feeling." We need to look at historical data to see what successful startups did rightâand what stalled startups did wrongâso we can invest our capital much smarter.

### 1.1 Context & Background

### 1.2 Core Analytical Objectives
-Speed & Momentum: Does moving fast actually matter? We'll check if the time it takes to get a first check, or the time between rounds, predicts long-term success.

-The Check Size: What's the funding sweet spot? Should we write a bunch of smaller checks or a few massive ones to give startups the best shot at advancing?

-The Right Markets: Which industries are consistently winning, and which ones are overcrowded and hitting a dead end?

-Location: Does where a startup is based genuinely change its odds of survival, or are traditional tech hubs overrated?

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
```

```python
# Load data using a relative path and Latin-1 encoding to handle special characters and creating a copy
startup=pd.read_csv('./DSB-FT_EDA Project/startup Investments/startup_investments.csv', encoding='latin1')
startup_cleaned= startup.copy()
```

## 3. Data Handling


```python
# Initial assessment: Check shapes, null values, and summary data types
print(f'The shape of the Dataset: \n{startup.shape}','\n\n\n')
print(f'The columns of the Dataset: {startup.columns}','\n\n\n')
print(f'The null sum of null values in the Dataset: {startup.isnull().sum()}','\n\n\n')
print(f'The data types of the Dataset: {startup.dtypes}')
```

```python
# Data cleaning operations (handling data types, text styling, missing values, etc.)
startup_cleaned.columns = startup_cleaned.columns.str.strip() 
#Strip any accidental leading or trailing whitespaces from the column names to prevent KeyErrors

#Removing commas then converting to float for the funding_total_usd
startup_cleaned['funding_total_usd']=startup_cleaned['funding_total_usd'].str.replace(',','',regex=True)
startup_cleaned['funding_total_usd']=pd.to_numeric(startup_cleaned['funding_total_usd'],errors='coerce')

startup_cleaned.sample(5)
```

```python
#Converting founded_at,first_funding_at, and last_funding_at from string to datetime
startup_cleaned['founded_at'] = pd.to_datetime(startup_cleaned['founded_at'], errors='coerce')
startup_cleaned['first_funding_at'] = pd.to_datetime(startup_cleaned['first_funding_at'], errors='coerce')
startup_cleaned['last_funding_at'] = pd.to_datetime(startup_cleaned['last_funding_at'], errors='coerce')

#Verify that the data types successfully shifted to datetime64[ns]
startup_cleaned[['founded_at', 'first_funding_at', 'last_funding_at']].dtypes
```

```python
#In this dataframe, there are over 4000 empty rows at the end, using .dropna() to remove them for a cleaner dataframe
startup_cleaned=startup_cleaned.dropna(how='all')
```

```python
#Modifying the cleaned startup dataframe to only keep the useful columns
columns_kept= [
    'name', 'status', 'market', 'category_list', 
    'country_code', 'state_code', 'region', 
    'funding_total_usd', 'funding_rounds', 
    'founded_at', 'first_funding_at', 'last_funding_at',
    'seed', 'venture', 'angel', 'round_A', 'round_B', 'round_C']

startup_cleaned=startup_cleaned[columns_kept]
```

```python
startup_cleaned.head()
```

```python
startup_cleaned[['status','market']].isnull().sum()
```

```python
#Clean missing values from critical columns, they cant be guessed or imputed safely
startup_cleaned = startup_cleaned.dropna(subset=['status', 'market'])
```

```python
#Checking for any remaining nulls if they need to be imputed or removed
startup_cleaned.isnull().sum()
```

```python
#there is 1 row with a missing name, lets imputed it with a name just to aviod errors down the line by finding out which row it is first
startup_cleaned[startup_cleaned['name'].isnull()]
```

```python
#generic name for the row with missing name
startup_cleaned['name'] = startup_cleaned['name'].fillna('Unknown Startup')
```

```python
# Handle country code, region, and state code nulls which are too high to just drop so fill them with unknown
startup_cleaned['country_code'] = startup_cleaned['country_code'].fillna('Unknown')
startup_cleaned['region'] = startup_cleaned['region'].fillna('Unknown')
startup_cleaned['state_code'] = startup_cleaned['state_code'].fillna('Non-US/Unknown')
```

```python
# 1. Calculate the sum of independent parent columns across each row
round_columns = startup_cleaned[['seed', 'angel', 'venture']].sum(axis=1)

#Fill the missing values in funding_total_usd using the calculated sum
startup_cleaned['funding_total_usd'] = startup_cleaned['funding_total_usd'].fillna(round_columns)

#Verify how many missing values remain now
print("Remaining missing funding totals:", startup_cleaned['funding_total_usd'].isnull().sum())
```

```python
#Lastly checking for any left missing data, other that founded which needs to be left alone
startup_cleaned.isnull().sum()
```

```python
#sampling the data just to check
startup_cleaned.sample(4)
```

```python
#Save the new cleaned dataframe to a fresh CSV file
startup_cleaned.to_csv('cleaned_startup_investments.csv', index = False)
startup_cleaned.head()
```

```python
#checking how many rows have seed captial of 0
startup_cleaned[startup_cleaned['seed'] == 0]
```

```python
#since the problem statment states that'allow the firm to accurately refine its $10M seed-stage investment criteria and target startups 
#with the highest odds of advancing.' isloating the startup targets that actually raises seed capital
seed_startup_cleaned= startup_cleaned[startup_cleaned['seed'] > 0].copy()

#From the problem statment 'analyze historical funding data across industry categories, funding round counts, time between rounds'
#Calculate the time intervals in days
seed_startup_cleaned['days_to_first_check']=(seed_startup_cleaned['first_funding_at'] - seed_startup_cleaned['founded_at']).dt.days
seed_startup_cleaned['time_between_rounds']=(seed_startup_cleaned['last_funding_at'] - seed_startup_cleaned['first_funding_at']).dt.days

# 3. Clean up the negative date anomalies (where investment was logged before incorporation)
seed_startup_cleaned=seed_startup_cleaned[seed_startup_cleaned['days_to_first_check'] >= 0]

```

```python
seed_startup_cleaned.isnull().sum()
```

```python
#Save the new cleaned dataframe to a fresh CSV file after refining it in accordance with the problem statment
seed_startup_cleaned.to_csv('seed_cleaned_startup.csv', index = False)
seed_startup_cleaned.head()
```

## 4. Exploratory Data Analysis (EDA)
*(Rubric Checkpoint: Demonstrate mastery of basic descriptive statistics, sorting records, and applying Boolean masking techniques to isolate specific baseline metrics.)*

### 4.1. Statistics Summary

```python
#summary statistics
print("--- Numerical Feature Distribution ---")
display(seed_startup_cleaned[['seed', 'funding_total_usd', 'funding_rounds', 'days_to_first_check', 'time_between_rounds']].describe())

#Baseline success vs. failure rates in the seed startup df
print("\n--- Startup Status Breakdown ---")
status_counts = seed_startup_cleaned['status'].value_counts()
status_perc = seed_startup_cleaned['status'].value_counts(normalize=True) * 100

display(pd.DataFrame({'Count': status_counts, 'Percentage (%)': status_perc}))
```

```python
#Bin the seed rounds into venture capital tiers
seed_bins = [0, 500000, 2000000, 5000000, float('inf')]
# Label each tier clearly so the financial numbers read intuitively for an investment committee
seed_labels = ['Micro(<$500k)', 'Standard ($500k-$2M)', 'Growth ($2M-$5M)', 'Mega (>$5M)']
seed_startup_cleaned['seed_bins'] = pd.cut(seed_startup_cleaned['seed'], bins=seed_bins, labels=seed_labels)

#Calculate the round A advancement rate for each tier
tier_analysis = seed_startup_cleaned.groupby('seed_bins', observed=False).agg(
    total_companies=('name', 'count'),
    raised_round_A=('round_A', lambda x: (x > 0).sum())
)
tier_analysis['round_A_success_rate_%'] = (tier_analysis['raised_round_A'] / tier_analysis['total_companies']) * 100
display(tier_analysis)
```

```python
#Isolate top 10 most common markets in our seed startup df to keep the analysis clean
top_markets = seed_startup_cleaned['market'].value_counts().head(10).index
# Slice out a subset dataframe containing only those 10 major competitive market categories
filtered_markets = seed_startup_cleaned[seed_startup_cleaned['market'].isin(top_markets)]

#Calculate exit/success rates by sector
market_success = filtered_markets.groupby('market').agg(
    total_startups=('name', 'count'),
    acquired_ipo=('status', lambda x: x.isin(['acquired', 'ipo']).sum())
)
market_success['success_rate_%'] = (market_success['acquired_ipo'] / market_success['total_startups']) * 100
display(market_success.sort_values(by='success_rate_%', ascending=False))
```

### 4.2. ----

```python
# Target Analysis Block 1: Numerical breakdown of funding velocity by venture outcome
velocity_summary = seed_startup_cleaned.groupby('status').agg(
    total_startups=('name', 'count'),
    median_days_to_check=('days_to_first_check', 'median'),
    average_days_to_check=('days_to_first_check', 'mean')
).sort_values(by='median_days_to_check')

display(velocity_summary)
#looking at the total_startups column for acquired if there are enough suxxessful exit examples to draw conclusions
```

### 4.3. ----

```python
# Visualization 1
#Establish presentation theme
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

#Build the distribution boxplot
#Using showfliers=False so extreme outliers don't compress the inner quartile boxes
#to check time between round change to this:
#data=seed_startup_cleaned[seed_startup_cleaned['time_between_rounds'] > 0]
#y='time_between_rounds'
sns.boxplot(
    data=seed_startup_cleaned,
    x='status',
    y='days_to_first_check',
    palette='Set2',
    hue='status',
    legend=False,
    showfliers=False 
)

#Formatting and labeling
plt.title('Funding Velocity: Days from Founding to First Seed Check by Venture Outcome', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Current Company Status', fontsize=12, fontweight='bold')
plt.ylabel('Days Elapsed Before First Seed Check', fontsize=12, fontweight='bold')

#Plotting
plt.tight_layout()
plt.show()
```

**Visualization 1 Interpretation:**
> *The boxplot chart above shows an unexpected relationship where failed startups in the closed status secure tehir first seed check faster with a median of 400 days. Compared to those still operating or acquired status both averaging a median of around 500 days. This directly reveals that intial seed funding is not reliable for long-term viability. For our $10M investment criteria, this indicates the firm should avoid over-indexing on immediate post-incorporation fundraising velocity, as a slightly longer operational runway before the first institutional check (around 1.3 to 1.5 years) correlates with significantly higher survival and positive exit trajectories.*

```python
# Visualization 2
#Round A Advancement Rate by Seed Check Size
plt.figure(figsize=(10, 5))

# 1. Define logical funding tiers for the seed check size
bins = [0, 500000, 1500000, 5000000, float('inf')]
labels = ['Micro (<$500k)', 'Standard ($500k-$1.5M)', 'Growth ($1.5M-$5M)', 'Mega (>$5M)']
seed_startup_cleaned['seed_bin'] = pd.cut(seed_startup_cleaned['seed'], bins=bins, labels=labels)

#Create a column indicating if they successfully raised a Series A
seed_startup_cleaned['raised_series_a'] = seed_startup_cleaned['round_A'] > 0

#Calculate the advancement rate percentages
advancement_data = seed_startup_cleaned.groupby('seed_bin', observed=False)['raised_series_a'].mean().reset_index()
advancement_data['success_rate_%'] = advancement_data['raised_series_a'] * 100

#Plotting
sns.barplot(
    data=advancement_data,
    x='seed_bin',
    y='success_rate_%',
    palette='crest',
    hue='seed_bin',
    legend=False
)

#Formatting and styling
plt.title('Round A Advancement Rate by Initial Seed Check Size', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Seed Funding Tier', fontsize=12, fontweight='bold')
plt.ylabel('Percentage Advancing to Round A (%)', fontsize=12, fontweight='bold')
plt.ylim(0, max(advancement_data['success_rate_%']) + 10)

plt.tight_layout()
plt.show()
```

**Visualization 2 Interpretation:**
This chart identifies the ideal capitalization window required to push a seed startup toward a follow-on Series A round. It reveals whether increasing initial check sizes structurally reduces the portfolio stagnation rate, or if returns on capital hit a point of diminishing returns. For our $10M fund allocation, this data gives us a strict boundary line to determine if we should optimize for high diversification (more micro-checks) or high conviction (fewer, larger checks).

```python
# Visualization 3
#Top 10 Sectors by Positive Exit Rate
plt.figure(figsize=(11, 6))

#Isolate the top 10 most common industries/markets in the seed cohort to keep the chart clean
top_10_markets = seed_startup_cleaned['market'].value_counts().head(10).index
market_df = seed_startup_cleaned[seed_startup_cleaned['market'].isin(top_10_markets)].copy()

#Define a positive exit as being currently 'acquired' or 'ipo'
market_df['is_successful_exit'] = market_df['status'].isin(['acquired', 'ipo'])

#Calculate the percentage of successful exits per market
market_success = market_df.groupby('market')['is_successful_exit'].mean().reset_index()
market_success['exit_rate_%'] = market_success['is_successful_exit'] * 100
market_success = market_success.sort_values(by='exit_rate_%', ascending=False)

#Plotting
sns.barplot(
    data=market_success,
    y='market',
    x='exit_rate_%',
    palette='viridis',
    hue='market',
    legend=False
)

#Formatting and styling
plt.title('Top 10 High-Density Seed Sectors Ranked by Successful Exit Rate (Acquired/IPO)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Successful Exit Rate (%)', fontsize=12, fontweight='bold')
plt.ylabel('Industry Market Sector', fontsize=12, fontweight='bold')
plt.xlim(0, market_success['exit_rate_%'].max() + 5)

plt.tight_layout()
plt.show()
```

**Visualization 3 Interpretation:**
This horizontal ranking isolates the highest-yielding sectors within the most common seed investment fields, separating genuinely lucrative startup ecosystems from overhyped, congested ones. By analyzing the structural variance in exit rates between sectors like software, biotechnology, or e-commerce, the chart establishes an empirical guide for industry sector weights. This allows our firm to strategically steer our $10M fund away from markets with low-realization ceilings and focus capital on historical exit leaders.

```python
#Visualization 4
#Round A Advancement Rate by Top 10 Regional Hubs
plt.figure(figsize=(11, 6))

#Filter out 'Unknown' countries and isolate the top 10 most active countries
known_countries = seed_startup_cleaned[seed_startup_cleaned['country_code'] != 'Unknown']
top_10_countries = known_countries['country_code'].value_counts().head(10).index
country_df = known_countries[known_countries['country_code'].isin(top_10_countries)].copy()

#Ensure our Series A success column is calculated
country_df['raised_round_a'] = country_df['round_A'] > 0
# 3. Calculate the advancement rate percentages per country
country_success = country_df.groupby('country_code')['raised_round_a'].mean().reset_index()
country_success['success_rate_%'] = country_success['raised_round_a'] * 100

#Create a clean dictionary mapping the codes to full names (add as many as you need)
country_name_map = {
    'ISR': 'No Name',
    'USA': 'United States',
    'DEU': 'Germany',
    'AUS': 'Australia',
    'GBR': 'United Kingdom',
    'CAN': 'Canada',
    'FRA': 'France',
    'ESP': 'Spain',
    'IND': 'India',
    'CHL': 'Chile'
}

#Apply the map to the country_code column
country_success['country_code'] = country_success['country_code'].map(country_name_map).fillna(country_success['country_code'])

#Sort the values *after* mapping so the bars stay in the perfect order
country_success = country_success.sort_values(by='success_rate_%', ascending=False)

#Generate the clean horizontal bar chart using the updated names
sns.barplot(
    data=country_success,
    y='country_code',  # This will now display the full names automatically!
    x='success_rate_%',
    palette='magma',
    hue='country_code',
   legend=False
)

#Apply institutional formatting and design tweaks
plt.title('Round A Advancement Rate Across Top 10 High-Density Countries', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Percentage Advancing to Round A (%)', fontsize=12, fontweight='bold')
plt.ylabel('Country Code', fontsize=12, fontweight='bold')
plt.xlim(0, country_success['success_rate_%'].max() + 5)

plt.tight_layout()
plt.show()
```

**Visualization 4 Interpretation:**
This chart scales our geographic analysis up to a macro level, ranking the top 10 countries by how effectively their seed-stage startups transition into a Series A capital round. By evaluating baseline success rates on a national scale, it highlights which cross-border regulatory or economic environments yield structurally more resilient startup cohorts. For our $10M fund distribution, this insight serves as a risk-mitigation framework to decide if our capital should remain heavily concentrated domestically or hedge into international markets with strong follow-on pipelines.

## 5. Summary of Findings & Actionable Recommendations
*(Guidelines: Connect individual observations into a cohesive narrative that guides the target non-technical audience toward data-driven actions.)*

### 5.1 Key Insights (Summary of Findings)
* Speed is a Trap: Our analysis showed a counterintuitive patternâfailed startups (closed) actually secured their seed checks faster (a median of ~400 days) than companies that went on to be acquired or remain stable (a median of ~500 days). Rushing an investment just because a team is fundraising immediately after incorporating is a statistically proven way to back a stalled company.

* The Funding Runway Matters: Startups that advance to Round A have a distinct capital "sweet spot." Teams that raise tiny micro-seed rounds often run out of money before hitting meaningful milestones, while mega-seeds don't automatically guarantee next-stage success. The data highlights a clear middle ground where capital size matches real operational runway.

* Avoid the Hype: Just because a market sector has a massive density of seed startups doesn't mean it yields actual returns. Some of the most crowded tech sectors suffer from incredibly low positive exit (Acquired/IPO) rates, proving they are oversaturated. Real value lies in high-yield, high-realization categories.

### 5.2 Actionable Recommendations
Based on these historical patterns, here is exactly how our firm should deploy and refine the $10M seed-stage investment criteria:
* Implement a "Maturity Mask" Filter: Do not rush to fund teams that are looking for institutional checks within their first 6â9 months of life. Instead, target founders who have bootstrapped or operated for roughly 1.3 to 1.5 years (the 500-day sweet spot). These teams historically display better resilience and product-market alignment.

* Optimize for Standard/Mid-Tier Conviction Checks: Rather than writing twenty tiny 500k dollars micro (which leaves startups undercapitalized and prone to stalling) or placing all 10M dollars on one or two mega-seeds, our fund should target a portfolio of 7 to 8 highly structured checks in the optimal tier shown in our advancement charts. This balances necessary runway with strategic diversification.

* Pivot Sector Weighting Toward Exit Leaders: Dynamically shift our target investment thesis away from low-conversion, overcrowded sectors. Even if a sector is highly popular in the media, our capital should prioritize industries with a proven track record of successful strategic acquisitions.
### 5.3 Limitations & Areas for Further Research
* The Micro-Funding Data Gap: Reconstructing the missing global funding values using only the primary pillars (seed, angel, venture) gave us a incredibly accurate baseline for tech startups. However, it intentionally overlooks alternative lifelines like bank debt, small business grants, or convertible notes, which might have helped a few stalled companies survive longer in the real world.

* Survival Blindspots from Drop Operations: Dropping rows missing both their status and market features early in our data handling pipeline was necessary to keep our target variables clean. However, it means our analysis cannot account for completely off-the-grid startups that might have wound down quietly without ever reporting an official "closed" status.

### 6.1 Data Dictionary
| Feature / Column | Data Type | Source | Description |
|:---|:---|:---|:---|
| name | Object | Original | The official operating name of the startup entity. Filled with 'Unknown Startup' where missing. |
| status | Object | Original | The current operational state of the company (`operating`, `acquired`, `closed`, `ipo`). Serves as a target variable. |

| founded_at | Datetime64 | Original | The official timestamp denoting when the company was incorporated or founded. |
| first_funding_at | Datetime64 | Original | The timestamp highlighting the company's first recorded institutional check or capital injection. |

| venture | Float64 | Original | Total institutional venture capital funding captured over the company's lifecycle in USD. |
| angel | Float64 | Original | Total angel investment funding raised from high-net-worth individuals in USD. |
| round_A | Float64 | Original | Total capital secured during the Series A institutional round in USD. |
| round_B | Float64 | Original | Total capital secured during the Series B institutional round in USD. |
| round_C | Float64 | Original | Total capital secured during the Series C institutional round in USD. |

| raised_series_a | Bool | Engineered | Boolean success flag indicating whether the startup successfully graduated to capture follow-on Series A funding (True / False). |
